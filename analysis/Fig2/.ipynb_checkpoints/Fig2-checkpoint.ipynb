```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

curr_dir = os.getcwd()

colors = sns.color_palette("crest", n_colors=6)

datasets = [
    {
        "mg": 0,
        "exp_file": "SASDFA9.dat",
        "scale": 3.0,
        "exp_color": "#b0d6bb",
        "fit_color": colors[0],
        "label": r"20 mM $\mathrm{Na^{+}}$",
    },
    {
        "mg": 1,
        "exp_file": "SASDFF9.dat",
        "scale": 1.5,
        "exp_color": "#bfdde5",
        "fit_color": colors[1],
        "label": r"1 mM $\mathrm{Mg^{2+}}$",
    },
    {
        "mg": 2,
        "exp_file": "SASDFG9.dat",
        "scale": 1.1,
        "exp_color": "#c5edee",
        "fit_color": colors[3],
        "label": r"2 mM $\mathrm{Mg^{2+}}$",
    },
    {
        "mg": 5,
        "exp_file": "SASDFH9.dat",
        "scale": 1.0,
        "exp_color": "#c6d5eb",
        "fit_color": colors[5],
        "label": r"5 mM $\mathrm{Mg^{2+}}$",
    },
]

fig, ax = plt.subplots(figsize=(5, 5))

for d in datasets:
    exp_path = os.path.join(
        curr_dir, "SAXS", "rA30", f"Mg{d['mg']}", d["exp_file"]
    )
    fit_path = os.path.join(
        curr_dir, "SAXS", "rA30", f"Mg{d['mg']}", "avg.dat"
    )

    exp = np.loadtxt(exp_path)
    fit = np.loadtxt(fit_path, skiprows=1)

    q_exp = exp[:, 0]
    I_exp = exp[:, 1]
    I_err = exp[:, 2]

    q_fit = fit[:, 0]
    I_fit = fit[:, 1]

    # Kratky transform: q^2 I(q)
    kratky_exp = I_exp * q_exp**2 * d["scale"]
    kratky_err = I_err * q_exp**2 * d["scale"]
    kratky_fit = I_fit * q_fit**2 * d["scale"]

    ax.errorbar(
        q_exp,
        kratky_exp,
        yerr=kratky_err,
        fmt="none",
        color=d["exp_color"],
        alpha=1,
        capsize=2,
        linewidth=1,
        zorder=1,
    )

    ax.scatter(
        q_exp,
        kratky_exp,
        color=d["exp_color"],
        alpha=1,
        s=30,
        zorder=2,
    )

    ax.plot(
        q_fit,
        kratky_fit,
        color=d["fit_color"],
        linewidth=3,
        label=d["label"],
        zorder=10,
    )

# Formatting
ax.set_xlim(0, 0.25)
ax.set_ylim(0, 0.35)

ax.set_xlabel(r"q ($\mathrm{\AA^{-1}}$)", fontsize=18)
ax.set_ylabel(r"$q^2 I(q)$ (a.u.)", fontsize=18)

ax.set_xticks([0.0, 0.1, 0.2])
ax.minorticks_on()

ax.tick_params(axis="both", which="major", length=5, labelsize=18)
ax.tick_params(axis="x", which="minor", length=3)

ax.set_yticks([])

ax.legend(
    frameon=False,
    fontsize=14,
    loc="best"
)

plt.tight_layout()
plt.show()
```

### Fig 2D.

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


def mean_ci_mbb(data, block_len=None, B=2000, confidence=0.95, rng=None):
    x = np.asarray(data, dtype=float)
    n = len(x)

    if n < 2:
        raise ValueError("Need at least 2 points.")

    if rng is None:
        rng = np.random.default_rng()

    if block_len is None:
        block_len = max(2, int(round(n ** (1 / 3))))

    n_blocks = int(np.ceil(n / block_len))
    boot_means = np.empty(B, dtype=float)

    X2 = np.concatenate([x, x[:block_len - 1]])

    for b in range(B):
        starts = rng.integers(0, n, size=n_blocks)
        y = np.concatenate([X2[s:s + block_len] for s in starts])[:n]
        boot_means[b] = y.mean()

    alpha = (1 - confidence) / 2
    lo, hi = np.quantile(boot_means, [alpha, 1 - alpha])

    mean = x.mean()
    return max(mean - lo, hi - mean)


base_dir = "/projects/academic/nguyenh/Peter/RNA-homopolymers/analysis-scripts/manuscript-final/Fig2"

mg = [0, 1, 2, 3, 5]
colors = sns.color_palette("deep")

rA_sasa_avg = []
rA_sasa_conf = []
rU_sasa_avg = []
rU_sasa_conf = []

rng = np.random.default_rng(12345)

for m in mg:
    rA_file = os.path.join(
        base_dir, "rA30", "Na20", f"Mg{m}", "sasa_partitioned_cg.dat"
    )

    rU_file = os.path.join(
        base_dir, "rU30", "Na20", f"Mg{m}", "sasa_partitioned_cg.dat"
    )

    rA_df = pd.read_csv(
        rA_file,
        comment="#",
        sep=r"\s+",
        names=["Base", "Sugar", "Phosphate", "Total"],
    )

    rU_df = pd.read_csv(
        rU_file,
        comment="#",
        sep=r"\s+",
        names=["Base", "Sugar", "Phosphate", "Total"],
    )

    rA_sasa_avg.append(np.average(rA_df["Phosphate"]))
    rA_sasa_conf.append(mean_ci_mbb(rA_df["Phosphate"], rng=rng))

    rU_sasa_avg.append(np.average(rU_df["Phosphate"]))
    rU_sasa_conf.append(mean_ci_mbb(rU_df["Phosphate"], rng=rng))


plt.figure(figsize=(5, 2.5))

plt.errorbar(
    mg,
    rA_sasa_avg,
    yerr=rA_sasa_conf,
    fmt="o-",
    capsize=10,
    markeredgecolor="white",
    color=colors[0],
    label=r"rA$_{30}$",
    lw=1.5,
    ms=10,
)

plt.errorbar(
    mg,
    rU_sasa_avg,
    yerr=rU_sasa_conf,
    fmt="o-",
    capsize=10,
    markeredgecolor="white",
    color=colors[3],
    label=r"rU$_{30}$",
    lw=1.5,
    ms=10,
)

plt.xlabel(r"$\mathrm{[Mg^{2+}]}$", fontsize=18)
plt.ylabel("P SASA (Å²)", fontsize=18)

plt.tick_params(axis="both", which="major", labelsize=18, direction="in")
plt.yticks(np.arange(2000, 2400, 100))
plt.xticks(mg)

plt.legend(frameon=False, fontsize=14)
plt.minorticks_on()
plt.tight_layout()
plt.show()
```
